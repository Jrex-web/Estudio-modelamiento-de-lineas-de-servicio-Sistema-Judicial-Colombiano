```python
#Distribucion de llegadas
#reporteconsolidado
import math
def distribucion_llegadas(fila):
    y=fila["Total_Entra_Mens"]
    t=fila["Meses_Reportados"]/fila["Meses_Reportados"] #=1
    n=70 
    me=y*t
    var=math.pow(me,n) 
    euler=math.e**-(me)
    factorialn=math.factorial(n)
    dp=(var/factorialn)*euler
    return dp
transformado["DistLleg_p/m"]=transformado.apply(distribucion_llegadas,axis=1)
reporteconsolidado["DistLleg_p/m"]=reporteconsolidado.apply(distribucion_llegadas,axis=1)
reporteconsolidado
```

```python
#Distribucion del tiempo medio de servicio
#expm1(1e-5)"Entr_TutImpug",

import math
def distribucion_tiempo_servicios_tutelas(fila):
    mu1=fila["Entr_TutImpug"]-fila["Egre_TutImpug"]
    mu2=fila["Entr_Acc_Const"]-fila["Egre_Acc_Const"]
    mut=mu1+mu2
    t=fila["Meses_Reportados"]/fila["Meses_Reportados"]
    e1=math.e**-mut*t
    e2=1-e1
    dm=e2
    return dm 
transformado["Distribución_MediaConst"]=transformado.apply(distribucion_tiempo_servicios_tutelas,axis=1)
reporteconsolidado["Distribución_MediaConst"]=reporteconsolidado.apply(distribucion_tiempo_servicios_tutelas,axis=1)
reporteconsolidado
```

```python
import math
def distribucion_tiempo_servicios_demanda(fila):
    m=fila["Total_Entra_Mens"]-fila["Egre_Prop"]
    t=fila["Meses_Reportados"]
    e1=math.e**-m*t
    e2=1-e1
    dm=e2
    return dm 
transformado["Distribución_MediaDem"]=transformado.apply(distribucion_tiempo_servicios_demanda,axis=1)
reporteconsolidado["Distribución_MediaDem"]=reporteconsolidado.apply(distribucion_tiempo_servicios_demanda,axis=1)
reporteconsolidado
```

```python
#Distribucion de la media,Tiempo Medio de Servicio varianza (1/m)2
def tiempo_medio(fila):
    tmu1=abs(1/fila["Egre_Prop"])#*0.2
    tmu2=abs(1/fila["Egre_TutImpug"])#*0.4
    tmu3=abs(1/fila["Egre_Acc_Const"])#*0.4
    tmt=tmu1+tmu2+tmu3
    return tmt 
transformado["Tiempo_Medio_Servicio"]=transformado.apply(tiempo_medio,axis=1)
reporteconsolidado["Tiempo_Medio_Servicio"]=reporteconsolidado.apply(tiempo_medio,axis=1)
reporteconsolidado
```

```python
#Tiempo Medio de Servicio varianza (1/m)2
def tiempo_medio_Varianza(fila):
    tmu1=abs(1/fila["Egre_Prop"])#*0.2
    tmu2=abs(1/fila["Egre_TutImpug"])#*0.4
    tmu3=abs(1/fila["Egre_Acc_Const"])#*0.4
    tmtvar=pow(tmu1,2)+pow(tmu2,2)+pow(tmu3,2)
    return tmtvar 
transformado["Tiempo_Medio_Servicio_varianza"]=transformado.apply(tiempo_medio_Varianza,axis=1)
reporteconsolidado["Tiempo_Medio_Servicio_varianza"]=reporteconsolidado.apply(tiempo_medio_Varianza,axis=1)
reporteconsolidado
```

```python
#Relación de Utilizacion promedio del sistema
#p = utilización promedio del sistema= Y/sm

def utilizacion_sistema_transformado(fila):
    y=fila["Total_Entra_Mens"]
    s=fila["Cantidad"]
    mu1=fila["Egre_Prop"]
    mu2=fila["Egre_TutImpug"]
    mu3=fila["Egre_Acc_Const"]
    mut=mu1+mu2+mu3
    us=s*mut
    ust=y/us
    return ust
def utilizacion_sistema_reporte(fila):
    y=fila["Total_Entra_Mens"]
    s=fila["Distrito"]
    mu1=fila["Egre_Prop"]
    mu2=fila["Egre_TutImpug"]
    mu3=fila["Egre_Acc_Const"]
    mu4=fila["Inv_Fin_Anual"]
    mut=mu1+mu2+mu3+mu4
    us=s*mut
    usc=y/us
    return usc

transformado["utilizacion_sistema"]=transformado.apply(utilizacion_sistema_transformado,axis=1)
reporteconsolidado["utilizacion_sistema"]=reporteconsolidado.apply(utilizacion_sistema_reporte,axis=1)
reporteconsolidado
```

```python
#Porcentaje de utilizacion del sistema
def porcentaje_utilizacion(fila):
    
    s=fila["Distrito"]
    y1=fila["Entr_Prop"]
    y2=fila["Entr_TutImpug"]
    y3=fila["Egre_Acc_Const"]
    yt=y1+y2+y3
    mu1=fila["Egre_Prop"]
    mu2=fila["Egre_TutImpug"]
    mu3=fila["Egre_Acc_Const"]
    mu4=fila["Inv_Fin_Anual"]
    mut=mu1+mu2+mu3+mu4
    ms=s*mut   
    pu1=yt/ms
    pu=abs(1-pu1)*100
    
    return pu
reporteconsolidado["Porcentaje_Utilizacion"]=reporteconsolidado.apply(porcentaje_utilizacion,axis=1)
reporteconsolidado
```

```python
"""
Probabilidad de 0 Clientes en el sistema
P0 = probabilidad de que cero clientes estén en el sistema
P0=∑_(n=0)^(s-1)▒[1/((γ/μ)^n/n!+〖(γ/μ)^s〗^ /s!*1/(1-p))] 
"""
import math
def sin_clientes(fila):
    n=0
    s=fila["Distrito"]
    y1=fila["Entr_Prop"]
    y2=fila["Entr_TutImpug"]
    y3=fila["Egre_Acc_Const"]
    yt=y1+y2+y3
    mu1=fila["Egre_Prop"]
    mu2=fila["Egre_TutImpug"]
    mu3=fila["Egre_Acc_Const"]
    mu4=fila["Inv_Fin_Anual"]
    mut=mu1+mu2+mu3+mu4
        
    p1=fila["Total_Entra_Mens"]
    p2=fila["Distrito"]
    p3=y1-mu1+y2-mu2+y3-mu3
    p4=p2*p3
    p=1/p4
    
    factorialn=math.factorial(n)
    factorials=math.factorial(s)
    
    p01=pow(yt/mut,n)/factorialn
    p02=pow(yt/mut,s)/factorials
    p03=1/(1-p)
    p04=p02*p03
    p05=p01+p04
    p0=1/p05
    #p0=(1/(pow(y1+y2+y3/mu1+mu2+mu3,n)/factorial1))+((pow(y1+y2+y3/mu1+mu2+mu3,s)/factorial2)*(1/1-p))
    return p0
reporteconsolidado["Prob_sinClientes"]=reporteconsolidado.apply(sin_clientes,axis=1)
reporteconsolidado
```

```python
#probabilidad n clientes
#Pn = probabilidad de que haya n clientes estén en el sistema =
def probabilidad_nclientes(fila):
    n=70
    s=fila["Distrito"]
    y1=fila["Entr_Prop"]
    y2=fila["Entr_TutImpug"]
    y3=fila["Egre_Acc_Const"]
    yt=y1+y2+y3
    mu1=fila["Egre_Prop"]
    mu2=fila["Egre_TutImpug"]
    mu3=fila["Egre_Acc_Const"]
    mu4=fila["Inv_Fin_Anual"]
    mut=mu1+mu2+mu3+mu4
    me=yt/mut

    p1=fila["Total_Entra_Mens"]
    p2=fila["Distrito"]
    p3=y1-mu1+y2-mu2+y3-mu3
    p4=p2*p3
    p=1/p4
    
    factorialn=math.factorial(n)
    factorials=math.factorial(s)
    
    p01=pow(me,n)/factorialn
    p02=pow(me,s)/factorials
    p03=1/(1-p)
    p04=p02*p03
    p05=p01+p04
    p0=1/p05
    pn1=p01
    pn2=n-s
    pn3=factorials*pow(s,pn2)
    pn4=p01/pn3
    pn=pn4*p0
    #pn1=(elevado/factorialn)*p0
    #pn2=(elevado/factorials)*p0
    return pn
reporteconsolidado["ProbVariosClientes"]=reporteconsolidado.apply(probabilidad_nclientes,axis=1)
reporteconsolidado
```

```python
#Promedio de clientes en fila de espera
#Lq = número promedio de clientes en la fila de espera =
def clientes_cola(fila):
    n=70
    s=fila["Distrito"]
    y1=fila["Entr_Prop"]
    y2=fila["Entr_TutImpug"]
    y3=fila["Egre_Acc_Const"]
    yt=y1+y2+y3
    mu1=fila["Egre_Prop"]
    mu2=fila["Egre_TutImpug"]
    mu3=fila["Egre_Acc_Const"]
    mut=mu1+mu2+mu3

    me=yt/mut

    p1=fila["Total_Entra_Mens"]
    p2=fila["Distrito"]/fila["Distrito"]
    p3=y1-mu1+y2-mu2+y3-mu3
    p4=p2*p3
    p=1/p4
        
    factorialn=math.factorial(n)
    factorials=math.factorial(s)
    
    p01=pow(me,n)/factorialn
    p02=pow(me,s)/factorials
    p03=1/(1-p)
    p04=p02*p03
    p05=p01+p04
    p0=1/p05
    
    lq1=p0*pow(me,s)*p
    lq2=1-p
    lq3=factorials*pow(lq2,2)
    lq4=lq1/lq3
    lq=1/lq4
    return lq
reporteconsolidado["Clientes en Cola"]=reporteconsolidado.apply(clientes_cola,axis=1)
reporteconsolidado
```

```python
#tiempo promedio de espera
# Wq = tiempo promedio de espera en la fila =
def tiempo_espera_cola(fila):
    n=70
    s=fila["Distrito"]
    y1=fila["Entr_Prop"]
    y2=fila["Entr_TutImpug"]
    y3=fila["Egre_Acc_Const"]
    yt=y1+y2+y3
    mu1=fila["Egre_Prop"]
    mu2=fila["Egre_TutImpug"]
    mu3=fila["Egre_Acc_Const"]
    mut=mu1+mu2+mu3

    me=yt/mut

    p1=fila["Total_Entra_Mens"]
    p2=fila["Distrito"]/fila["Distrito"]
    p3=y1-mu1+y2-mu2+y3-mu3
    p4=p2*p3
    p=1/p4
        
    factorialn=math.factorial(n)
    factorials=math.factorial(s)
    
    p01=pow(me,n)/factorialn
    p02=pow(me,s)/factorials
    p03=1/(1-p)
    p04=p02*p03
    p05=p01+p04
    p0=1/p05
    
    lq1=p0*pow(me,s)*p
    lq2=1-p
    lq3=factorials*pow(lq2,2)
    lq4=lq1/lq3
    lq=1/lq4
    wq=lq/yt
    return wq
reporteconsolidado["Tiempo_Espera"]=reporteconsolidado.apply(tiempo_espera_cola,axis=1)
reporteconsolidado
```

```python
#Tiempo en el sistema
# W = tiempo promedio transcurrido en el sistema, incluido el servicio = Wq+1/m

def tiempo_sistema(fila):
    n=70
    s=fila["Distrito"]
    y1=fila["Entr_Prop"]
    y2=fila["Entr_TutImpug"]
    y3=fila["Egre_Acc_Const"]
    yt=y1+y2+y3
    mu1=fila["Egre_Prop"]
    mu2=fila["Egre_TutImpug"]
    mu3=fila["Egre_Acc_Const"]
    mu4=mu4=fila["Inv_Fin_Anual"]
    mut=mu1+mu2+mu3+mu4

    me=yt/mut

    p1=fila["Total_Entra_Mens"]
    p2=fila["Distrito"]/fila["Distrito"]
    p3=y1-mu1+y2-mu2+y3-mu3
    p4=p2*p3
    p=1/p4
        
    factorialn=math.factorial(n)
    factorials=math.factorial(s)
    
    p01=pow(me,n)/factorialn
    p02=pow(me,s)/factorials
    p03=1/(1-p)
    p04=p02*p03
    p05=p01+p04
    p0=1/p05
    
    lq1=p0*pow(me,s)*p
    lq2=1-p
    lq3=factorials*pow(lq2,2)
    lq4=lq1/lq3
    lq=1/lq4
    lq3=factorials*pow(lq2,2)
    lq3=factorials*pow(lq2,2)
    wq=lq/yt  
    w1=wq+1/mut
    w=w1*100                                                                   
    return w

reporteconsolidado["Tiempo_sistema"]=reporteconsolidado.apply(tiempo_sistema,axis=1)
reporteconsolidado
```

```python
#tiempo clientes en lsistema
#L = tiempo promedio de clientes en el sistema de servicio = y*W
def tiempo_clientes_sistema(fila):
    n=70
    s=fila["Distrito"]
    y1=fila["Entr_Prop"]
    y2=fila["Entr_TutImpug"]
    y3=fila["Egre_Acc_Const"]
    yt=y1+y2+y3
    mu1=fila["Egre_Prop"]
    mu2=fila["Egre_TutImpug"]
    mu3=fila["Egre_Acc_Const"]
    #mu4=fila["Inv_Fin_Anual"]#/fila["Distrito"]
    mut=mu1+mu2+mu3#+mu4

    me=yt/mut

    p1=fila["Total_Entra_Mens"]
    p2=fila["Distrito"]/fila["Distrito"]
    p3=yt-mut#y1-mu1+y2-mu2+y3-mu3+mu4
    p4=p2*p3
    p=1/p4
        
    factorialn=math.factorial(n)
    factorials=math.factorial(s)
    
    p01=pow(me,n)/factorialn
    p02=pow(me,s)/factorials
    p03=1/(1-p)
    p04=p02*p03
    p05=p01+p04
    p0=1/p05
    
    lq1=p0*pow(me,s)*p
    lq2=1-p
    lq3=factorials*pow(lq2,2)
    lq4=lq1/lq3 
    lq=1/lq4    #Numero de clientes en cola
    wq=lq/yt    #Tiempo promedio en cola
    w1=wq+1/mut
    w=w1   #Tiempo transcurido en el sistmea
    l=yt*w #Clientes atendidos en el sistema durante el periodo promedio de cioentes en lsistema
    return l
reporteconsolidado["clientes/Mes"]=reporteconsolidado.apply(tiempo_clientes_sistema,axis=1)
reporteconsolidado
```
